```python
df['cut'] = df['cut'].replace({'Ideal': 0, 'Premium': 1, 'Good': 2, 'Very Good': 3, 'Fair': 4})
df['color'] = df['color'].replace({'E': 0, 'I': 1, 'J': 2, 'H': 3, 'F': 4, 'G': 5, 'D': 6})
# Define the mapping
clarity_mapping = {'SI2': 0, 'SI1': 1, 'VS1': 2, 'VS2': 3, 'VVS2': 4, 'VVS1': 5, 'I1': 6, 'IF': 7}

# Apply the mapping
df['clarity'] = df['clarity'].replace(clarity_mapping)
```

```python
from sklearn.model_selection import train_test_split

#split data
X = df[['carat']]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

#create and train the  linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error(MAE): {mae}")
print(f"Mean Squared Error(MSE): {mse}")
print(f"R-squared: {r2}")
```
