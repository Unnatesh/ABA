```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier 
```

```python
## Probability practical
#a simulate rolling die
```

```python
# simulation
num_rolls = 100000
outcomes = np.random.randint(1, 7, num_rolls)
outcomes
```

```python
#calculate probabilites
unique, counts = np.unique(outcomes, return_counts=True)
probabilities = counts / num_rolls
print(probabilities)
```

```python
# display results
for value, prob in zip(unique, probabilities):
    print(f"Outcome: {value} : Probability = {prob:.4f} ")
```

```python
#plot histogram
plt.figure(figsize = (8, 5))
sns.barplot(x=unique, y = probabilities, palette="viridis")
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.title("Probability Distribution of Rolling a Die")
plt.show()
```

```python
# b. generate random numbers from distribution
n = 100000
normal_data = np.random.normal(loc = 50, scale = 10, size = n) #Mean = 50, Std Dev = 10
uniform_data = np.random.uniform(low = 0, high = 100, size = n) # range 0 to 100
exponential_data = np.random.exponential(scale= 10, size = n) #mean = 10
```

```python
#plot histogram
fig, ax = plt.subplots(1, 3, figsize=(18, 5))


def generate_histplot(data, title, x, color):
    sns.histplot(data, bins = 50, kde = True, ax=ax[x], color = color)
    ax[x].set_title(title)

generate_histplot(normal_data, "Normal Distribution (Mean=50, std=10)", 0, "blue")
generate_histplot(uniform_data, "Uniform Distribution (Range=0 to 100)", 1, "red")
generate_histplot(exponential_data, "Exponential Distribution (Mean = 10)", 2, "green")

```
